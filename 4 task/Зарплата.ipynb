```python
class Purchase:
    def __init__(self, rate, days):
        self.__mrot = 13890
        self.__rate = rate
        if 1 <= days < 32:
            self.__days = days
        else:
            raise ValueError
        self.__oklad = 8 * self.__days * self.__rate
        self.__zp = self.__oklad * 2
        self.__zp_s_uk = self.__zp * 1.15
        self.__ndfl = self.__zp_s_uk * 0.15
        self.__zp_s_ndfl = self.__zp_s_uk - self.__ndfl

    def CulcResults(self):
        self.__oklad = 8 * self.__days * self.__rate
        self.__zp = self.__oklad * 2
        self.__zp_s_uk = self.__zp * 1.15
        self.__ndfl = self.__zp_s_uk * 0.15
        self.__zp_s_ndfl = self.__zp_s_uk - self.__ndfl

    def __Result__(self):
        if self.__zp_s_ndfl > self.mrot:
            return "Заработная плата больше прожиточного минимума"
        elif self.__zp_s_ndfl < self.mrot:
            return "Заработная плата меньше прожиточного минимума"
        else:
            return "Заработная плата равна прожиточному минимуму"

    def __str__(self):
        return f"Оклад = {round(self.__oklad, 2)}\n" \
               f"Премия = {round(self.__zp, 2)}\n" \
               f"ЗП с урал.коэфф. = {round(self.__zp_s_uk,  2)}\n" \
               f"НДФЛ = {round(self.__ndfl, 2)}\n" \
               f"ЗП после вычета НДФЛ = {round(self.__zp_s_ndfl, 2)}\n" \
               f"{self.__Result__()}"

    def __repr__(self):
        return f"Оклад = {round(self.__oklad, 2)}\n" \
               f"Премия = {round(self.__zp, 2)}\n" \
               f"ЗП с урал.коэфф. = {round(self.__zp_s_uk,  2)}\n" \
               f"НДФЛ = {round(self.__ndfl, 2)}\n" \
               f"ЗП после вычета НДФЛ = {round(self.__zp_s_ndfl, 2)}\n" 

    @property
    def rate(self):
        return self.__rate

    @rate.setter
    def rate(self, rate):
        self.__rate = rate

    @property
    def days(self):
        return self.__days

    @days.setter
    def days(self, days):
        if 1 <= days <= 31:
            self.__days = days
        else:
            raise ValueError

    @property
    def mrot(self):
        return self.__mrot

    @mrot.setter
    def mrot(self, mrot):
        self.__mrot = mrot

    def get_oklad(self):
        return self.__oklad

    def get_zp(self):
        return self.__zp

    def get_zp_s_uk(self):
        return self.__zp_s_uk

    def get_zp_s_ndfl(self):
        return self.__zp_s_ndfl

while True:
    rate=int(input('Введите часовую тарифную ставку: '))
    days=int(input('Введите количество рабочих дней: '))
    if 1 <= days < 32:
        break
    print("Введите правильное количество дней!")
    
p = Purchase (rate, days)
print(str(p))
```
